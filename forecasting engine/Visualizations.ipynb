```python
df = pd.read_csv("data\State_Processed.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
for state in df['state'].unique():
    state_data = df[df['state'] == state]
    plt.plot(state_data['month_date_yyyymm'], state_data['median_listing_price'], label=state)

plt.xlabel('month_date_yyyymm')
plt.ylabel('median_listing_price')
plt.title('Median Listing Price Over Time for Each State')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(14, 8))

# Loop through regions
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    
    # Aggregate median listing price across states in the region
    region_monthly = region_data.groupby('month_date_yyyymm')['median_listing_price'].mean()
    
    plt.plot(region_monthly.index, region_monthly.values, label=region)

plt.xlabel('Month')
plt.ylabel('Median Listing Price')
plt.title('Median Listing Price Over Time by Region')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(14, 8))

# Loop through regions
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    
    # Aggregate average listing price across states in the region
    region_monthly = region_data.groupby('month_date_yyyymm')['average_listing_price'].mean()

    plt.plot(region_monthly.index, region_monthly.values, label=region)

plt.xlabel('Month')
plt.ylabel('Average Listing Price')
plt.title('Average Listing Price Over Time by Region')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(14, 8))

# Loop through divisions
for division in df['Division'].unique():
    division_data = df[df['Division'] == division]

    # Aggregate average listing price across states in the division
    division_monthly = division_data.groupby('month_date_yyyymm')['average_listing_price'].mean()

    plt.plot(division_monthly.index, division_monthly.values, label=division)

plt.xlabel('Month')
plt.ylabel('Average Listing Price')
plt.title('Average Listing Price Over Time by Division')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
